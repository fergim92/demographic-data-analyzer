### 1. `head(n=5)`: Devuelve las primeras n filas de la serie.
### 2. `tail(n=5)`: Devuelve las últimas n filas de la serie.
### 3. `describe()`: Calcula estadísticas descriptivas de la serie, como la media, la mediana, etc.
### 4. `unique()`: Devuelve un array con los valores únicos de la serie.
### 5. `value_counts()`: Devuelve un conteo de ocurrencias de cada valor único en la serie.
### 6. `count()`: Devuelve el número de elementos no nulos en la serie.
### 7. `sum()`: Devuelve la suma de todos los elementos de la serie.
### 8. `mean()`: Devuelve la media de la serie.
### 9. `median()`: Devuelve la mediana de la serie.
### 10. `std()`: Devuelve la desviación estándar de la serie.
### 11. `min()`: Devuelve el valor mínimo de la serie.
### 12. `max()`: Devuelve el valor máximo de la serie.
### 13. `idxmin()`: Devuelve el índice del valor mínimo de la serie.
### 14. `idxmax()`: Devuelve el índice del valor máximo de la serie.
### 15. `sort_values()`: Ordena los valores de la serie.
### 16. `sort_index()`: Ordena la serie por sus índices.
### 17. `isin(values)`: Devuelve una serie booleana que indica si cada elemento está en el conjunto de valores proporcionado.
### 18. `fillna(value)`: Rellena los valores faltantes (NaN) con el valor especificado.
### 19. `dropna()`: Elimina los valores faltantes (NaN) de la serie.
### 20. `map(func)`: Aplica una función a cada elemento de la serie.
### 21. `apply(func)`: Aplica una función a la serie.
### 22. `astype(dtype)`: Convierte el tipo de datos de la serie al tipo especificado.
### 23. `str`: Proporciona acceso a una serie de métodos de cadenas para manipulación de texto.

# Data Frames

## Metodos mas comunes

Claro, a continuación, te listo algunos de los métodos más comunes que tienen los DataFrames en pandas, junto con breves descripciones de su funcionalidad:

### 1. `head(n=5)`: Devuelve las primeras n filas del DataFrame.
### 2. `tail(n=5)`: Devuelve las últimas n filas del DataFrame.
### 3. `shape`: Devuelve una tupla con las dimensiones (filas, columnas) del DataFrame.
### 4. `info()`: Proporciona un resumen de información sobre el DataFrame, como tipos de datos y valores no nulos.
### 5. `describe()`: Calcula estadísticas descriptivas para todas las columnas numéricas del DataFrame.
### 6. `columns`: Devuelve una lista con los nombres de las columnas del DataFrame.
### 7. `index`: Devuelve los índices del DataFrame.
### 8. `dtypes`: Devuelve los tipos de datos de las columnas del DataFrame.
### 9. `isna()`: Devuelve un DataFrame booleano que indica si cada elemento es nulo (NaN).
### 10. `dropna()`: Elimina filas que contienen al menos un valor nulo (NaN).
### 11. `fillna(value)`: Rellena los valores nulos (NaN) con el valor especificado.
### 12. `astype(dtype)`: Convierte el tipo de datos de las columnas al tipo especificado.
### 13. `sort_values(by=column, ascending=True)`: Ordena el DataFrame por los valores de una columna específica.
### 14. `sort_index()`: Ordena el DataFrame por sus índices.
### 15. `groupby(by=column)`: Agrupa el DataFrame por los valores de una columna específica.
### 16. `sum()`: Devuelve la suma de los valores de las columnas del DataFrame.
### 17. `mean()`: Devuelve la media de los valores de las columnas del DataFrame.
### 18. `median()`: Devuelve la mediana de los valores de las columnas del DataFrame.
### 19. `max()`: Devuelve el valor máximo de cada columna del DataFrame.
### 20. `min()`: Devuelve el valor mínimo de cada columna del DataFrame.
### 21. `count()`: Devuelve el número de valores no nulos en cada columna del DataFrame.
### 22. `unique()`: Devuelve una lista con los valores únicos de una columna específica.
### 23. `value_counts()`: Devuelve un conteo de ocurrencias de cada valor único en una columna específica.
### 24. `apply(func)`: Aplica una función a lo largo de las columnas del DataFrame.
### 25. `rename(columns=dict)`: Cambia los nombres de las columnas del DataFrame usando un diccionario.
### 26. `drop(columns=columnas)`: Elimina columnas específicas del DataFrame.
### 27. `merge(df2, how='inner', on=column)`: Combina dos DataFrames utilizando una columna común.
### 28. `concat([df1, df2])`: Concatena dos o más DataFrames en uno solo.
### 29. `pivot(index, columns, values)`: Crea una tabla dinámica (pivot table) a partir del DataFrame.
### 30. `to_csv(file_path)`: Guarda el DataFrame como un archivo CSV en la ubicación especificada.
